```python
wind_current= pd.read_csv('ninja_wind_europe_v1.1_current_national.csv', index_col='time')

# Converting the index as date
wind_current.index = pd.to_datetime(wind_current.index)
```

```python
wind_current.AT.mean() # Ok ça colle bien avec Excel 'Meta Data for Wind Farms'
```

```python
wind_current.DE.mean() # Ok ça colle bien avec Excel 'Meta Data for Wind Farms'
```

```python
yearly_wind = wind_current.groupby(wind_current.index.year).mean() 
yearly_wind.head()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12,5))

yearly_wind[['DE','FR','NL','DK']].plot(ax=ax, title='Capacity factor for 4 countries')

ax.set_xlabel('Year')
ax.set_ylabel('Capacity factor')
```
